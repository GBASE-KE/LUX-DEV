## WE IMPORT OUR DATA

```python
import pandas as pd
```

```python
malaria = pd.read_excel("Malaria Cases and Testing 2021.xlsx")
```

```python
malaria.head(10)
```

```python
#define agg functions
combined_counties = malaria.groupby(["County"]).aggregate({'jan':sum, 'feb':sum, 'mar':sum, 'apr':sum, 'may':sum, 'jun':sum, 'jul':sum,
       'aug':sum, 'sep':sum, 'oct':sum, 'nov':sum, 'dec':sum})


combined_counties.head(10)
```

## Visualizing the data

```python
import matplotlib as plt
```

```python
def plot(x,y, format="-r", lineLabel="", xLabel="",yLabel="", title=""):
    fig, ax=plt.subplots()
    ax.plot(x,y, label=lineLabel)
    ax.legend(loc=2)
    ax.set_xlabel(xLabel)
    ax.set_ylabel(yLabel)
    ax.set_xticks(x)
      
```

```python
y=list(combined_counties.iloc[0])
y=y[2:-1]
```

```python
x=range(1,13)
plt.rcParams['figure.figsize']=(7,6)
plot(x,y,lineLabel=r"Month vs Cases", title="Baringo 2021",  xLabel="month",yLabel="value")
plt.show() 
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```
